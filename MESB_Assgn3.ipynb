```python
A = 100
VmA = 10
KmA = 100
B = 0
C= 0
VmB = 30
KmB = 900
a = []
b = []
c = []
while True:
    ViA = (VmA*A)/(KmA+A)
    A = A - ViA
    B = B + ViA
    if A < 0.0000001:
        break
    if B > 0.1:
        ViB = (VmB*B)/(KmB+B)
        B = B - ViB
        C = C +  ViB
    else:
        pass
    a.append(A)
    b.append(B)
    c.append(C)
    #print(B)
l=[]
for i in range(len(a)):
    l.append(i)

import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(l,a)
ax.plot(l,b)
ax.plot(l,c)


    
```
